```python
import os
import yaml

# 1. Find the dataset folder under /content
root = '/content'
folders = [f for f in os.listdir(root) if os.path.isdir(os.path.join(root, f))]
print("Folders in /content:", folders)

# Assume the last folder is your dataset (usually named like projectname-1)
dataset_folder = folders[-1]
dataset_path = os.path.join(root, dataset_folder)
print("\n‚ûú Dataset path:", dataset_path)

# 2. Show subfolders under the dataset path
print("\nContents of dataset folder:", os.listdir(dataset_path))

# 3. Show train/valid/test counts
for split in ['train', 'valid', 'test']:
    img_dir = os.path.join(dataset_path, split, 'images')
    if os.path.isdir(img_dir):
        print(f"{split} images:", len(os.listdir(img_dir)))
    else:
        print(f"{split} folder NOT found at", img_dir)

# 4. Read data.yaml if it exists
data_yaml = os.path.join(dataset_path, 'data.yaml')
if os.path.isfile(data_yaml):
    print("\nFound data.yaml. Here are its contents:")
    with open(data_yaml) as f:
        config = yaml.safe_load(f)
    print("  nc (number of classes):", config.get('nc'))
    print("  Class names:")
    for idx, name in enumerate(config.get('names', [])):
        print("   ", idx, "‚Üí", name)
else:
    print("\nNo data.yaml found in dataset folder.")
```

```python
# CELL 2: Import Libraries
from ultralytics import YOLO
from roboflow import Roboflow
from google.colab import drive
import os
import yaml
from IPython.display import Image, display

print("‚úÖ Libraries loaded successfully!")
```

```python
# CELL 3: Mount Google Drive (to save your trained model)
print("üíæ Connecting to Google Drive...")
drive.mount('/content/drive', force_remount=True)

# Create project directory
project_dir = '/content/drive/MyDrive/dental_ai_pro'
os.makedirs(project_dir, exist_ok=True)
print(f"‚úÖ Project directory: {project_dir}")
```

```python
# CELL 4: Download Dataset from Roboflow
print("üì• Downloading dental dataset...")

rf = Roboflow(api_key="s8Tm6H8avg4YN9e4I1PY")
project = rf.workspace("abdo-pnblp").project("dental-caries-i8vaj-rath1")
version = project.version(1)
dataset = version.download("yolov8")

print(f"‚úÖ Dataset downloaded to: {dataset.location}")
```

```python
# CELL 5: Configure Dataset
data_yaml_content = f"""
path: {dataset.location}
train: train/images
val: valid/images
test: test/images

nc: 1
names:
  0: caries
"""

data_yaml_path = os.path.join(dataset.location, 'data.yaml')
with open(data_yaml_path, 'w') as f:
    f.write(data_yaml_content)

print("‚úÖ Dataset configuration complete!")
print(f"Configuration file: {data_yaml_path}")
```

```python
# CELL 6: Load Pre-trained YOLOv8 Model
print("ü§ñ Loading YOLOv8 model...")
model = YOLO('yolov8m.pt')  # medium model
print("‚úÖ Model loaded: yolov8m (medium)")
```

```python
# CELL 7: Train Your Model
print("üöÄ Starting training... this may take 1-3 hours")
results = model.train(
    data=data_yaml_path,
    epochs=100,
    imgsz=640,
    batch=16,
    patience=15,
    save=True,
    device=0,
    project=project_dir,
    name='dental_detection_v1',
    optimizer='AdamW',
    lr0=0.001,
    weight_decay=0.0005,
    warmup_epochs=5,
    cos_lr=True,
    augment=True,
    val=True,
    plots=True,
    save_period=10
)
print("üéâ Training Complete!")
print("Best model at:", f"{project_dir}/dental_detection_v1/weights/best.pt")
```

```python
# CELL 8: Display Training Results
print("üìä Training Results:")
display(Image(filename=f'{project_dir}/dental_detection_v1/results.png'))
display(Image(filename=f'{project_dir}/dental_detection_v1/confusion_matrix.png'))
```

```python
# CELL 9: Validate Your Model
print("üîç Validating model performance...")
best_model = YOLO(f'{project_dir}/dental_detection_v1/weights/best.pt')
validation_results = best_model.val(
    data=data_yaml_path,
    split='test',
    plots=True
)
print(f"\nPrecision: {validation_results.box.mp:.3f}")
print(f"Recall: {validation_results.box.mr:.3f}")
print(f"mAP@0.5: {validation_results.box.map50:.3f}")
print(f"mAP@0.5-0.95: {validation_results.box.map:.3f}")

if validation_results.box.map50 >= 0.90:
    print("‚úÖ EXCELLENT - Clinical grade")
elif validation_results.box.map50 >= 0.85:
    print("‚úÖ GOOD - Screening")
elif validation_results.box.map50 >= 0.80:
    print("‚ö†Ô∏è ACCEPTABLE - More data")
else:
    print("‚ùå NEEDS IMPROVEMENT")
```

```python
# CELL 10: Test on Sample Image
print("üñºÔ∏è Testing on sample image...")
test_images = os.listdir(f"{dataset.location}/test/images")
sample_image = f"{dataset.location}/test/images/{test_images[0]}"
results = best_model.predict(
    source=sample_image,
    conf=0.5,
    save=True,
    project=project_dir,
    name='test_predictions'
)
display(Image(filename=f'{project_dir}/test_predictions/{test_images[0]}'))
print("‚úÖ Sample detection complete")
```
